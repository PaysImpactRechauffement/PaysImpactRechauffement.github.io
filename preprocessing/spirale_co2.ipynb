```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

data = pd.read_csv(
    "co2.csv",
    index_col=["year", "month"]
)

data = data.loc[1850:]
data = data.drop(['datenum', 'datetime', 'day'], axis=1)
data.head()

data.plot()

data.loc[2010:].plot()
```

```python
data = data.reset_index()

data.rename(columns={"data_mean_global": "value"}).to_csv("concentrations.csv", index=False)
data.rename(columns={"data_mean_nh": "value"}).to_csv("concentrations_nh.csv", index=False)
data.rename(columns={"data_mean_sh": "value"}).to_csv("concentrations_sh.csv", index=False)
```
